```python
import pandas as pd
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import scipy
```

```python
exc=pd.read_excel('r1.xlsx')
```

```python
x=exc['highest']
y=exc['lowest']
```

```python

parameter = np.polyfit(x, y, 1)
y2=parameter[0]*x+parameter[1]
correlation = np.corrcoef(y, y2)[0,1]  #relative coefficient
r2=correlation**2   #R2
plt.figure(figsize=(8,6),dpi=150)
plt.ylim(0, 120)
plt.plot(x, y2, color='k')
for index,row in exc.iterrows():
     plt.scatter(x=int(row['highest']),y=int(row['lowest']),marker=row['marker'],c=row['color'],label=row['name'])
plt.xlabel('The highest evelation of the top of dam (m)')
plt.ylabel('The lowest evelation of the top of dam (m)')
plt.text(80,20,s=r'$y={:.2f}x+{:.2f} (R^2={:.3})$'.format(parameter[0],parameter[1],r2))
plt.legend()
plt.show()
```
